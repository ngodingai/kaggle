# Data Cleaning
methods: rename columns, fillna, astype

```python
median_penjualan = df['Penjualan'].median()
median_hs = df['Harga Satuan'].median()
```

```python
median_penjualan, median_hs
```

```python
df['Penjualan'] = df['Penjualan'].fillna(median_penjualan)
df['Harga Satuan'] = df['Harga Satuan'].fillna(median_hs)
```

```python
df
```

# Modify Dataframe
new columns, apply function

```python
df['Total'] = df['Penjualan'] * df['Harga Satuan']
```

```python
df
```

# Sorting Data

# Merge and Concat

```python
df_arhan = df[df['Operator']=='Arhan']
df_sari = df[df['Operator']=='Sari']
```

```python
df_arhan
```

```python
df_sari
```

```python
df_arhan_sari = pd.concat([df_arhan, df_sari])
```

```python
df_arhan_sari
```

```python
df_arhan_clean = df_arhan[['Tanggal', 'Total']].rename(columns={
    'Total': 'Total_Arhan'
})
```

```python
df_sari_clean = df_sari[['Tanggal', 'Total']].rename(columns={
    'Total': 'Total_Sari'
})
```

```python
df_arhan_clean_agg = df_arhan_clean.groupby(['Tanggal']).sum().reset_index()
```

```python
df_sari_clean_agg = df_sari_clean.groupby(['Tanggal']).sum().reset_index()
```

```python
df_arhan_clean_agg
```

```python
df_sari_clean_agg
```

```python
df_final = pd.merge(df_arhan_clean_agg, df_sari_clean_agg, left_on='Tanggal', right_on='Tanggal', how='inner')
```
