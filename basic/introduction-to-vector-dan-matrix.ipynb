```python
import numpy as np
```

```python
# Definisi vektor
a = np.array([1, 2, 3, 5, 6])
b = np.array([4, 5, 6, 3, 4])

print("Vektor a:", a)
print("Vektor b:", b)
```

```python
a.shape, b.shape
```

## Dot Product


```python
dot_product = np.dot(a, b)
print("Dot product:", dot_product)
```

```python
dot_product_2 = a@b
print("Dot product:", dot_product_2)
```

## Cosine Similarity


```python
cos_sim = dot_product / (np.linalg.norm(a) * np.linalg.norm(b))
print("Cosine similarity:", cos_sim)
```

# Matrix

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("Matriks A:\n", A)
print("Matriks B:\n", B)
```

```python
A.shape, B.shape
```

```python
C = A + B
print("A + B =\n", C)
```

```python
D = np.dot(A, B)   # atau A @ B
print("A * B =\n", D)
```

```python
A_T = A.T
print("Transpose A:\n", A.T)

```

```python
A_T
```

```python
A_inv = np.linalg.inv(A)
print("Inverse dari A:\n", A_inv)

```

```python
# verifikasi (A * A_inv = Identity)
print("A @ A_inv =\n", np.round(A @ A_inv, 2))
```
